Missing data found in Age and Salary column

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy ='mean')
imputer.fit(x[:, 1:3])
x[:, 1:3] = imputer.transform(x[:, 1:3])
```

Replacing missing data with the mean value of the column

```python
print(x)
```

Missing values have been transformed

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[("encoder",OneHotEncoder(),[0])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

We need to transform the features column string data into binaries and also we need to convert x to numpy arrays before we can use it for training. For encoding strings to binaries we used OneHotEncoder here.

```python
print(x)
```

Here we can see our COUNTRY column has changed into binary values for our machine to work with it.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

We need to transform the labels column string data into binaries and we don't need to convert y to numpy arrays before we can use it for training. For encoding label strings to binaries we used LabelEncoder here.

```python
print(y)
```

Here we can see our PURCHASED column has changed into binary values for our machine to work with it.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

Here we are feature scaling to bring all the data to a Standard Scale for our machine.

```python
print (x)
```

As we can see all the datas are now in a standard range for our machine understand and use them for training and testing.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size = 0.2, random_state = 0)
```

Here we are splitting our feature matrices and dependent variables in training and test set by setting 20% for the test set size and 80% for the training set size otherwise overfitting may occur if the test size is higher.
